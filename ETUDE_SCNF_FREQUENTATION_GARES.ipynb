```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.colors as mcolors


data = pd.read_csv('frequentation-gares-vf.csv')

data['Log Total Voyageurs 2022'] = np.log1p(data['Total Voyageurs 2022'])


plt.figure(figsize=(10, 8))


norm = mcolors.PowerNorm(gamma=2, vmin=data['Log Total Voyageurs 2022'].min(), vmax=data['Log Total Voyageurs 2022'].max())


scatter = plt.scatter(data['Longitude'], data['Latitude'], c=data['Log Total Voyageurs 2022'], cmap='plasma_r', norm=norm, s=20)


plt.colorbar(scatter, label='Fréquentation 2022 (échelle exponentielle)')


plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.xlim(-5, 10)
plt.ylim(42, 51)


plt.title('Fréquentation des Gares en 2022 par Ville ')


plt.show()
```
